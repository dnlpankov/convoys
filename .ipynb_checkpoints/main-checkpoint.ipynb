```python

df1=df0.dropna(subset=['first_contact_date'])[['batch_number']].copy()
df1['batch_number']=df1['batch_number'].astype('category')
df1['first_contact_date']=pd.to_datetime(df0.first_contact_date, format='%d/%m/%y %H:%M').dt.date
df1['conversion_date']=pd.to_datetime(df0.agreement_reached_date, format='%d/%m/%y %H:%M').dt.date
df1['now']='29/01/10 17:14'
df1['now']=pd.to_datetime(df1['now']).dt.date
```

```python
import convoys.utils
unit, groups, (G, B, T) = convoys.utils.get_arrays(
    df1, groups='batch_number', created='first_contact_date', converted='conversion_date',
    unit='days', group_min_size=100)
```

```python
import convoys.plotting
import matplotlib.pyplot as plt
convoys.plotting.plot_cohorts(G, B, T, model='kaplan-meier', ci=0.95, groups=groups, t_max=60)
plt.legend()
plt.show()
```
